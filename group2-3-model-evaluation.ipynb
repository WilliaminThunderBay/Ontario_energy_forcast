```python
import numpy as np
import matplotlib.pyplot as plt

def plot_classification_metrics_radar(class_dict):
    """
    Plots a radar chart for classification metrics:
    Accuracy, Precision, Recall, F1-score
    
    class_dict should be in the format:
      {
        'ModelName': [accuracy, precision, recall, f1_score],
        'AnotherModel': [...],
        ...
      }
    """
    # Define the metric labels
    metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1-score']
    num_metrics = len(metrics_names)

    # Create the angles for each metric
    angles = np.linspace(0, 2 * np.pi, num_metrics, endpoint=False)
    # Close the circle by appending the start angle at the end
    angles = np.concatenate((angles, [angles[0]]))

    # Initialize the figure and polar subplot
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, polar=True)
    ax.set_title("Classification Metrics (Radar Chart)", y=1.1, fontsize=14)

    # Plot each model's metrics
    for model_name, values in class_dict.items():
        # Close the circle by repeating the first value at the end
        stats = values + [values[0]]
        ax.plot(angles, stats, label=model_name)
        ax.fill(angles, stats, alpha=0.25)

    # Set up the axes & labels
    ax.set_thetagrids(angles[:-1] * 180 / np.pi, metrics_names)
    ax.set_ylim(0, 1.2)  # Adjust upper limit if needed
    ax.grid(True)
    ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1.05))
    plt.show()


# Example usage:
# Suppose you already have a dictionary of classification metrics:
class_metrics = {
    'LightGBM':    [0.718, 0.718, 0.901, 0.758],
    'RandomForest':[0.831, 0.818, 0.884, 0.850],
    'XGBoost':     [0.928, 0.903, 0.903, 0.950],
    'ARIMA':       [0.541, 0.541, 1.000, 0.702]
}

plot_classification_metrics_radar(class_metrics)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------------
# 1) Define Your Evaluation Metrics
# --------------------------------------------------
# For each model: [MSE, RMSE, MAE, R^2]
# We'll ignore the 4th element (R^2) for plotting.

demand_metrics = {
    'LightGBM':    [1.51171912506157e12, 1.22951988e6, 9.2960048e5,  0.728],
    'RandomForest':[1.60965709328245e12, 1.26872262e6, 9.5346515e5,  0.710],
    'XGBoost':     [1.71320043110400e12, 1.30889283e6, 9.8586169e5,  0.692],
    'ARIMA':       [3.3640897776182186e36, 1.8341455170237225e18, 1.1067713386591612e18, -6.055372244665612e20]
}

price_metrics = {
    'LightGBM':    [1137.24, 33.72, 18.81,  -0.091],
    'RandomForest':[1137.66, 33.73, 19.24,  -0.092],
    'XGBoost':     [1735.49, 41.66, 22.57,  -0.666],
    'ARIMA':       [6.640474190492018e27, 8.148910473487862e13, 4.892969381735805e13, -6.372962959469402e27]
}

# Classification metrics: [Accuracy, Precision, Recall, F1-score]
class_metrics = {
    'LightGBM':    [0.718, 0.718, 0.901, 0.758],
    'RandomForest':[0.831, 0.818, 0.884, 0.850],
    'XGBoost':     [0.928, 0.903, 0.903, 0.950],
    'ARIMA':       [0.541, 0.541, 1.000, 0.702]
}

# Confusion matrices for each model
confusion_matrices = {
    'LightGBM':    np.array([[9998, 3102], [1522, 13915]]),
    'RandomForest':np.array([[10070, 3030],[1788, 13649]]),
    'XGBoost':     np.array([[9678, 3422], [1493, 13944]]),
    'ARIMA':       np.array([[5, 13095],   [0, 15437]])
}





def plot_confusion_matrices(cm_dict):
    """
    Plots each confusion matrix in a grid of subplots.
    """
    num_models = len(cm_dict)
    rows = int(np.ceil(num_models / 2))
    fig, axs = plt.subplots(rows, 2, figsize=(10, 5*rows))
    axs = axs.flatten()

    for ax, (model, cm) in zip(axs, cm_dict.items()):
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax)
        ax.set_title(f"{model} Confusion Matrix")
        ax.set_xlabel("Predicted")
        ax.set_ylabel("Actual")
        ax.grid(False)
    
    # Hide any unused subplots if num_models is odd
    if num_models < len(axs):
        for j in range(num_models, len(axs)):
            axs[j].set_visible(False)
    
    plt.tight_layout()
    plt.show()

plot_confusion_matrices(confusion_matrices)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1) Example Regression Metrics
# -----------------------------
demand_metrics = {
    'LightGBM':    [1.51171912506157e12, 1.22951988e6, 9.2960048e5,  0.728],
    'RandomForest':[1.60965709328245e12, 1.26872262e6, 9.5346515e5,  0.710],
    'XGBoost':     [1.71320043110400e12, 1.30889283e6, 9.8586169e5,  0.692],
    'ARIMA':       [3.3640897776182186e36, 1.8341455170237225e18, 1.1067713386591612e18, -6.055372244665612e20]
}

price_metrics = {
    'LightGBM':    [1137.24, 33.72, 18.81,  -0.091],
    'RandomForest':[1137.66, 33.73, 19.24,  -0.092],
    'XGBoost':     [1735.49, 41.66, 22.57,  -0.666],
    'ARIMA':       [6.640474190492018e27, 8.148910473487862e13, 4.892969381735805e13, -6.372962959469402e27]
}

# -----------------------------
# 2) Log-Transform Function
#    (Ignoring R^2, the 4th element)
# -----------------------------
def log_transform_three_metrics(data_dict):
    """
    Returns a dict { model: [log10(MSE), log10(RMSE), log10(MAE)] }.
    Ignores the 4th element (R^2).
    """
    log_metrics = {}
    for model, values in data_dict.items():
        mse, rmse, mae, _ = values
        log_mse  = np.log10(mse)  if mse  > 0 else None
        log_rmse = np.log10(rmse) if rmse > 0 else None
        log_mae  = np.log10(mae)  if mae  > 0 else None
        log_metrics[model] = [log_mse, log_rmse, log_mae]
    return log_metrics

demand_log_metrics = log_transform_three_metrics(demand_metrics)
price_log_metrics  = log_transform_three_metrics(price_metrics)

# -----------------------------
# 3) Two-Subplot Heatmap Function
# -----------------------------
def plot_regression_heatmaps(demand_log_dict, price_log_dict):

    metrics_labels = ['log10(MSE)', 'log10(RMSE)', 'log10(MAE)']
    models_demand = list(demand_log_dict.keys())
    data_demand   = np.array([demand_log_dict[m] for m in models_demand], dtype=float)

    models_price = list(price_log_dict.keys())
    data_price   = np.array([price_log_dict[m] for m in models_price], dtype=float)
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    sns.heatmap(data_demand, annot=True, fmt=".2f", cmap="YlGnBu",
                xticklabels=metrics_labels, yticklabels=models_demand, ax=axes[0])
    axes[0].set_title("Demand - log10(MSE, RMSE, MAE)")

    sns.heatmap(data_price, annot=True, fmt=".2f", cmap="YlGnBu",
                xticklabels=metrics_labels, yticklabels=models_price, ax=axes[1])
    axes[1].set_title("Price - log10(MSE, RMSE, MAE)")

    plt.tight_layout()
    plt.show()

# -----------------------------
# 4) Generate the Two-Subplot Figure
# -----------------------------
plot_regression_heatmaps(demand_log_metrics, price_log_metrics)
```
